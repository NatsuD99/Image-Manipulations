```
import matplotlib.pyplot as plt
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```
img=cv2.imread('clown.jpg')
img_grey=cv2.imread('clown.jpg', 0)
cv2_imshow(img)
cv2_imshow(img_grey)
```

```
def calc_hist(img):
    hgram = cv2.calcHist([img],[0],None,[256],[0,256])
    return hgram
```

```
def plot_hist(hist1, hist2, first='Color', second='Greyscale'):
 
    fig, (ax1, ax2)=plt.subplots(1, 2, figsize=(12, 4))
    ax1.plot(hist1)
    ax1.set_title(f'{first} image histogram', fontdict={'color': 'white'})
    ax2.plot(hist2)
    ax2.set_title(f'{second} image histogram', fontdict={'color': 'white'})
    plt.tight_layout()
    plt.show()
```

```
hist1=calc_hist(img)
hist2=calc_hist(img_grey)
plot_hist(hist1, hist2)
```

```
equ=cv2.equalizeHist(img_grey)
hist3=calc_hist(equ)
```

```
plot_hist(hist2, hist3, first='Greyscale', second='Equalized')
```

```
res = np.hstack((img_grey,equ))
cv2_imshow(res)
```
